# 9.1.1 Specifying the data model

Das Datenmodell (Likelihood) muss nun flexibler gemacht werden. Einerseits muss $\mu_i$ abhängig von X (`temp`) gemacht werden und auch die Variabilität muss lernbar sein:

$$
Y_i | \beta_0, \beta_1, \sigma \stackrel{\text{ind}}{\sim} N(\mu_i, \sigma^2) \quad \text{with} \quad \mu_i = \beta_0 + \beta_1 X_i
$$

Wird eine Lineare regression durchgeführt, so werden folgende annahmen angenommen:
1. Die Fehlerterme sind nicht autokorreliert
2. Die Beziehung zwischen X und Y lässt sich durch eine lineare Funktion modelieren
3. Die Variabilität der Fehlerterme sind normalverteilt um $\mu_i$ über den ganzen Definitionsbereich

# 9.1.2 Specifying the priors

$$
\beta_0 \sim N(m_0, s_0^2)
$$

$$
\beta_1 \sim N(m_1, s_1^2)
$$

$$
\sigma \sim \text{Exp}(l).
$$


Annahmen:
- Alle Parameter sind voneinander unabhängig

# 9.1.3 Putting it all together

$$
\begin{aligned}
\text{data:} \quad & Y_i | \beta_0, \beta_1, \sigma \stackrel{\text{ind}}{\sim} N(\mu_i, \sigma^2) \quad \text{with} \quad \mu_i = \beta_0 + \beta_1 X_i \\
\text{priors:} \quad & \beta_0 \sim N(m_0, s_0^2) \\
& \beta_1 \sim N(m_1, s_1^2) \\
& \sigma \sim \text{Exp}(l).
\end{aligned}
$$

# 9.2 Tuning prior models for regression parameters

Wir haben folgende Annahmen aus dem Buch:

1. On an average temperature day, say 65 or 70 degrees for D.C., there are typically around 5000 riders, though this average could be somewhere between 3000 and 7000.

2. For every one degree increase in temperature, ridership typically increases by 100 rides, though this average increase could be as low as 20 or as high as 180.

3. At any given temperature, daily ridership will tend to vary with a moderate standard deviation of 1250 rides.

```python
# intercept
plot_normal_prior(mean = 5000, sd = 1000)

# slope
plot_normal_prior(mean = 100, sd=40)

# sigma
plot_gamma_prior(1, 0.0008)
```

```python
priors = pd.DataFrame({
    "Intercept": stats.norm(5000 - 100*70, 100).rvs(200), # Calculate the intercept with slope of 100 and the fact that the average number of riders is 5000 at an average day with temperature of 70°
    "Slope": stats.norm(100, 40).rvs(200),
    "Sigma": stats.gamma(1, 1/0.0008).rvs(200)
})

priors
```

```python
x = np.linspace(45, 90, 1000)

for i, prior in priors.iterrows():
    y = prior.Intercept + prior.Slope*x

    plt.plot(x, y, color="black", alpha=0.4)

plt.title("200 simulated model lines under the specified priors")
plt.xlabel("temperature")
plt.ylabel("number of riders")
plt.show()
```

# 9.3 Posterior simulation

Wie sehen die Daten aus:

```python
sns.scatterplot(Bikeshare, x="temperature", y="count")
```

Nun möchten wir das ganze simulieren:

$$
f(\beta_0, \beta_1, \sigma \mid \vec{y}) \propto \text{prior} \cdot \text{likelihood} = f(\beta_0) f(\beta_1) f(\sigma) \cdot \left[ \prod_{i=1}^{n} f(y_i \mid \beta_0, \beta_1, \sigma) \right].
$$

Berechnen wir nun die Posterior:

$$
f(\beta_0, \beta_1, \sigma \mid \vec{y}) =
\frac{\text{prior} \cdot \text{likelihood}}{\int \text{prior} \cdot \text{likelihood}}
$$

$$
= \frac{ f(\beta_0) f(\beta_1) f(\sigma) \cdot \left[ \prod_{i=1}^{n} f(y_i \mid \beta_0, \beta_1, \sigma) \right] }
{\int \int \int f(\beta_0) f(\beta_1) f(\sigma) \cdot \left[ \prod_{i=1}^{n} f(y_i \mid \beta_0, \beta_1, \sigma) \right] d\beta_0 d\beta_1 d\sigma}
$$

So fällt auf, dass dies nicht wirklich ausgerechnet werden kann. Deshalb wird es simuliert:

# 9.3.1 Simulation via Bambi

```python
# define prior distributions as above
priors = {
    "Intercept": bmb.Prior("Normal", mu=5000, sigma=1000),
    "temperature": bmb.Prior("Normal", mu=100, sigma=40),
    "Sigma": bmb.Prior("Exponential", scale=0.0008)
}

# disable auto scaling (uninformative prior); disable center_predictors
results, fit, model = fit_glm("count ~ temperature", Bikeshare, auto_scale=False, center_predictors=False, return_all=True, priors=priors)
```

# 9.4 Interpreting the posterior

Nun möchten wir den posterior interpretieren:

```python
mcmc_summarize(results)
```

```python
results_combined = mcmc_combine_chains(results).pivot(index=["sample_nr", "is_warmup", "chain"], columns="var_name", values="value").reset_index()
results_combined.columns.name = None

results_combined.head()
```

```python
x = np.linspace(0, 1, 10)
for i, row in results_combined.iloc[:50].iterrows():
    y = row.Intercept + x*row.temperature
    plt.plot(x, y, alpha=0.1, color="black", label="bayes linred" if i == 49 else None)

sns.scatterplot(Bikeshare, x="temperature", y="count", alpha=0.1, label="data points")
sns.regplot(Bikeshare, x="temperature", y="count", scatter=False, color="red", label="reference linreg")
plt.legend()

plt.title("Bayes Linear Regression vs least squares linear regression")
plt.show()
```

Möchten wir noch herausfinden, ob der positive zusammenhang statistisch signifikant ist, so können wir die wahrscheinlichkeit dafür angeben:

```python
proba_mcmc(results, variable="temperature", pi_l_than=0)
```

Wie man sehen kann, ist die Wahrscheinlichkeit dafür null. Somit lässt sich mit einer hohen sicherheit sagen, dass ein positiver zusammenhang besteht!

# 9.5 Posterior prediction

Wie wir vom letzten Kapitel noch wissen, setzt sich die Unsicherheit aus der Sampling variability und der Posterior variability zusammen.

Dies analytisch zu rechnen kann man gerade vergessen. Deshalb muss dies mit den resultaten der mcmc erledigen:

```python
temp_pred = np.array([0, 0.2, 0.4, 0.6, 0.8, 1]) # which points do we want to predict?

preds = []
for data_point in temp_pred:
    mu = results_combined.Intercept + data_point*results_combined.temperature
    
    # create a sample with the new mean and sigma
    sample = stats.norm(mu, results_combined.sigma).rvs()

    preds.append(pd.DataFrame({
        data_point: sample
    }))

preds = pd.concat(preds, axis=1)

# make long
preds = preds.melt(var_name="temperature", value_name="count")

preds.head()
```

```python
x = np.linspace(0, 1, 10)
for i, row in results_combined.iloc[:50].iterrows():
    y = row.Intercept + x*row.temperature
    plt.plot(x, y, alpha=0.1, color="black", label="bayes linred" if i == 49 else None)

sns.scatterplot(Bikeshare, x="temperature", y="count", alpha=0.1, label="data points")
sns.scatterplot(preds, x="temperature", y="count", alpha=0.1)
plt.legend()

plt.title("Bayes Linear Regression with predictions")
plt.show()
```
